```python
# This function add two numbers 
def add(a,b):
    return a + b

#This function substracts two numbers
def substract(a,b):
    return a - b

# This function multiplies two numbers 
def multiply(a,b):
    return a * b 
# This function Divides two numbers
def divide(a,b):
    return a/b

print("select operation.")
print("1.add")
print("2.substract")
print("3.multiply")
print("4.divide")

while True:
    choice = input("Enter choice(1/2/3/4):")

    # Check if choice is one of the four options
    if choice in ('1','2','3','4'):
        try:
          num1 = int(input("Enter first number."))
          num2 = int(input("enter second number."))
        except ValueError:
            print("invalid input.please enter a number.")
            continue 
        if choice == '1':
            print(num1, "+", num2, "=",
                  add(num1,num2))
        elif choice == '2':
            print(num1, "-", num2, "=",
                  substract(num1,num2))
        elif choice == '3':
            print(num1, "*", num2, "=",
                  multiply(num1,num2))
        elif choice == '4':
            print(num1, "/", num2, "=",
                  divide(num1,num2))

            #if user wants another calculation 
            next_calculation = input("Let's do next calculation?(yes/no):")
            if next_calculation == "no":
                break
            else:
               print("Invalid Input")


```

```python
#Task 2: Design a ROCK-PAPER-SCISSORS GAME using Python Programming.

import random

 

options = ["Rock", "Paper", "Scissors"]

 

user_choice = input("Choose Rock, Paper, or Scissors: ")

computer_choice = random.choice(options)

 

print("You chose: ", user_choice)

print("Computer chose: ", computer_choice)

 

if user_choice == computer_choice:

    print("It's a tie!")

elif user_choice == "Rock" and computer_choice == "Scissors":

    print("You win!")

elif user_choice == "Paper" and computer_choice == "Rock":

    print("You win!")

elif user_choice == "Scissors" and computer_choice == "Paper":

    print("You win!")

else:

    print("Computer wins!")
```

```python
# Task - 3 : Design a "TO DO LIST" by using python programming.

# Import the necessary modules
import tkinter as tk

# Create the main window
window = tk.Tk()
window.title("To-Do List")

# Create a frame to hold the to-do list items
frame = tk.Frame(window)
frame.pack()

# Create a listbox to display the to-do list items
listbox = tk.Listbox(frame)
listbox.pack()

# Create a function to add a new item to the to-do list
def add_item():
    item = entry.get()
    listbox.insert(tk.END, item)
    entry.delete(0, tk.END)

# Create a function to remove a selected item from the to-do list
def remove_item():
    listbox.delete(tk.ACTIVE)

# Create a function to mark a selected item as complete
def complete_item():
    listbox.itemconfig(tk.ACTIVE, foreground="gray")

# Create a button to add a new item
add_button = tk.Button(frame, text="Add", command=add_item)
add_button.pack()

# Create a button to remove a selected item
remove_button = tk.Button(frame, text="Remove", command=remove_item)
remove_button.pack()

# Create a button to mark a selected item as complete
complete_button = tk.Button(frame, text="Complete", command=complete_item)
complete_button.pack()

# Create an entry widget to allow the user to enter a new item
entry = tk.Entry(frame)
entry.pack()

# Start the mainloop
window.mainloop()
```
